```
HS_records = list(studentData['High School']) 
print(HS_records) 
highSchool_dict = {i: HS_records.count(i) for i in HS_records} 
print(highSchool_dict)
```

```
# Pie Chart Time! 

# Setting up parameters

params = {'legend.fontsize': 15,
          'legend.handlelength': 1.5}
plt.rcParams.update(params)

labels = highSchool_dict.keys()
frequencies = highSchool_dict.values()
colors = ['gold', 'silver','blue', 'red', 'green', 'purple', 'yellow', 'orange'] 

# Time to plot 
plt.figure(figsize=(14,8))
plt.pie(frequencies, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140) 
plt.axis("equal")
plt.legend()
```

```
# Let's get the frequency of of the different grade levels 

print(studentData['Grade'].value_counts())

# Generating a dictionary of the grade levels and their frequencies 

gradeCounts = dict(studentData['Grade'].value_counts()) 
print(gradeCounts)
plt.figure(figsize=(14,8))

# Histogram

studentData['Grade'].hist(color="green") 
plt.xlabel("Grade Level", size="24")
plt.ylabel("Frequency", size="24")
plt.title("Distribution of Grade Levels for M132 Students", size="40")
plt.tight_layout()
```

```

# Let's see the counts for the gender distribution 

studentData['Gender'].value_counts() 

# Notice how close the female number is to male? That's extremely impressive for a data science class. Just saying :D 

# What else did you notice?
# Getting proportions of each gender 

## First build list comprehension from value count dictionary 

genderCounts = list(dict(studentData['Gender'].value_counts()).values()) 
genderKeys = list(dict(studentData['Gender'].value_counts()).keys())
print(genderCounts) 
print(genderKeys)

# Now let's get the proportions

print(f"Male: {round((genderCounts[0] / sum(genderCounts)) * 100)}% \n \
Female: {round((genderCounts[1] / sum(genderCounts)) * 100)}%")
```
