# Model

```python
from scipy import stats
```

```python
pred_price = stats.linregress(x=df.sqft_living,y=df.price)
```

```python
pred_price
```

```python
def predict(x):
    return pred_price.slope * x + pred_price.intercept
```

```python
predict(5400)
```
